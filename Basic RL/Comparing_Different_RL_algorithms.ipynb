# **Comparing Different RL algorithms**

```python
from Q_learning import Q_Learning
from SARSA import SARSA_learning
from Sophisticated_Q_learning import Q_Learning_Randomness
from RL_utils import find_path, calculate_rewards, path_in_grids, average_R_display, steps_plot, plot_Q_changing
import numpy as np
import matplotlib.pyplot as plt
```

## **Q-Learning**

```python
q_learning = Q_Learning()
q_learning.max_episodes = 10000
q_learning.rewards={'r_time':-0.5,'r_pond':-5,'r_croissant':30,'r_cogs':20,'r_work':20}
q_learning.tubes=[[(0, 0), (1, 5)], [(1, 2), (4, 1)]]
q_learning.epsilon_decays={'epsilon_threshold': 0.5, 'epsilon_decay1': 0.9999, 'epsilon_decay2': 0.999}
```

```python
%%time
a_hist_ql, Q_hist_ql, Rtot_ql = q_learning.learn(0.8,0.9,0.9)
```

```python
plot_Q_changing(Q_hist_ql, absolute=True, rolling_periods=200)
```

```python
plot_Q_changing(Q_hist_ql, absolute=True)
```

```python
average_R_display(Rtot_ql, rolling_periods=200)
```

```python
average_R_display(Rtot_ql)
```

```python
steps_plot(a_hist_ql, rolling_periods=200)
```

```python
steps_plot(a_hist_ql)
```

```python
path = find_path(Q_hist_ql[-1])
r, steps = calculate_rewards(q_learning.R, path)
print("Path: ",path)
print("Path in grid coords: ", path_in_grids(path, q_learning.dims))
print("Total Rewards: ", r)
print("Number of steps: ",steps)
```
